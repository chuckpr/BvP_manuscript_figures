```python
%%R
library(phyloseq)
library(dplyr)
library(ggplot2)
```

```python
%%R
physeq.bact = import_biom("../data/B/otu_table.wtax_wsd.biom")
physeq.bact.snorm = transform_sample_counts(physeq.bact, function(x) x / sum(x))
mdf.bact = tbl_df(psmelt(physeq.bact.snorm))
```

```python
%%R
p.ranks = mdf.bact %.% 
            filter(EnvironmentType == "planktonic") %.% 
            group_by(Sample) %.%
            mutate(rank = rank(-Abundance, ties.method = "first")) %.%
            select(OTU, CarbonToPhosphorous, Date, rank)
mdf.bact.wRanks = left_join(mdf.bact, p.ranks, by = c("Date", "OTU", "CarbonToPhosphorous"))
```

```python
%%R
?left_join
```

```python
%%R

#d = mdf.bact.wRanks %.% group_by(Sample) %.% arrange(rank) #%.% filter(EnvironmentType == "biofilm")
d = mdf.bact.wRanks %.% filter(EnvironmentType == "biofilm")

p = ggplot(mdf.bact.wRanks, aes(x = rank, y = Abundance, color = EnvironmentType))

p = p + geom_line()

p = p + facet_grid(CarbonToPhosphorous ~ Date)

p = p + scale_y_log10()

p = p + scale_x_log10()

p = p + labs(x = "OTU Ranked by Planktonic Abundances\n in Each C:P/Date Group")

p
```
