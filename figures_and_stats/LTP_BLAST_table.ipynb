```python
%%R
library(data.table)
library(dplyr)
```

```python
%%R
blastout_dt = fread("../data/tmp/blast_out_otusn_LTP115.m6")
blastout_df = tbl_df(blastout_dt)
colnames(blastout_df) = c("OTU",
                        "acc",
                        "pid",
                        "aln.length",
                        "mismatches",
                        "gapOpenCount",
                        "qstart",
                        "qend",
                        "sstart",
                        "send",
                        "e.value",
                        "bit.score")

top_hits_df_LTP = blastout_df %>% 
                          group_by(OTU) %>% 
                          mutate(Rank = rank(desc(bit.score), ties.method = "min")) %>% 
                          filter(Rank == 1)
```

```python
%%R
dt = fread("/var/seq_data/LTP115_db/LTPs115_SSU.csv", sep = ";")
colnames(dt) = c("acc","start","end","DB.version","name","unknown1","family","unknown2","url","full_tax")
LTP_seq_data = tbl_df(dt)
```

```python
%%R
joined_df = left_join(top_hits_df_LTP, LTP_seq_data)
```

```python
%%R
df.l2fc = read.table("../data/BvP.l2fc.bact.csv", sep = ",", header = TRUE, stringsAsFactors = FALSE)
df.l2fc.summary = df.l2fc %>% group_by(OTU) %>% summarize(m = max(abs(log2FoldChange)) * sign(log2FoldChange), 
                                                          Rank2 = first(Rank2), 
                                                          Rank6 = first(Rank6),
                                                          Rank3 = first(Rank3),
                                                          Rank5 = first(Rank5),
                                                          minP = min(padj)) %>% arrange(desc(abs(m)))
```

```python
%%R
head(df.l2fc.summary)
```

```python
%%R
N = 25
FDR = 0.10
topN = arrange(df.l2fc.summary, desc(abs(m))) %>% filter(minP <= FDR) %>% select(OTU, Rank2, m)
topN_otus = topN$OTU[1:N]
topN_df = left_join(joined_df, topN) %>% filter(OTU %in% topN_otus) %>% arrange(desc(abs(m)))
```

```python
df = com.load_data("topN_df")
```

```python
def get_latex(df):
    latex = "" 
    prefix = r"\multirow{%s}{*}{%s}"%(len(df.index), df.OTU.iloc[0]) \
           + r" & \multirow{%s}{*}{%s}"%(len(df.index), df.Rank2.iloc[0]) \
           + r" & \multirow{%s}{*}{%0.2f}"%(len(df.index), df.m.iloc[0])
    c = 0 
    for i, r in df[["name","pid","acc"]].iterrows():
        if c == 0:
            latex += " & " + " & ".join([str(x) for x in r.values]) + r" \\ "
        else:
            latex += " & & & " + " & ".join([str(x) for x in r.values]) + r" \\ "
        c += 1
    return prefix + latex + r"\midrule" + "\n"
```

```python
s = df.groupby("OTU").apply(get_latex)

table_framework = r"""
\documentclass[11pt]{article}
\usepackage{multirow, array, booktabs}
\begin{document}

\thispagestyle{empty}
 
\begin{table}[ht]
\caption{Results for BLAST search against Living Tree Project (top 25 environment enriched OTUs)}
\begin{center}
\begin{tabular}{llr>{\itshape}lrl}
    \toprule \\
    \textbf{OTU ID} & \textbf{Phylum} & $log_2(plankton:biofilm)$ & \textbf{Species Name} & \textbf{BLAST percent identity} & \textbf{accession} \\
    \midrule
    %s
    \bottomrule
\end{tabular}
\end{center}
\label{tab:multicol}
\end{table}
 
\end{document}"""%"".join(s.values)

with open("../data/tmp/table.tex", "w") as out:
    out.write(table_framework)
```
