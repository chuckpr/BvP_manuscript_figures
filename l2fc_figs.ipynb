```python
%%R
library(ggplot2)
library(RColorBrewer)
```

```python
%%R -w 800 -h 400
DF = read.table("../data/BvP.l2fc.bact.csv", 
                sep = ",",
                header = TRUE)

df.plot = subset(DF, !is.na(DF$padj))

colourCount = length(unique(df.plot$Rank2))
getPalette = colorRampPalette(brewer.pal(5, "Dark2"))

df.top = data.frame(x = df.plot$Rank4, y = -log(df.plot$padj), OTU = df.plot$OTU,
                    panel = "-log(p-value)", c = df.plot$Rank2)
df.bottom = data.frame(x = df.plot$Rank4, y = df.plot$log2FoldChange, OTU = df.plot$OTU,
                       panel = "log[2](planktonic:biofilm)", 
                       c = df.plot$Rank2)

d = rbind(df.top, df.bottom)
d.bact = d

p = ggplot(data = d, mapping = aes(x = x, y = y))
p = p + geom_hline(y = 0) 
p = p + facet_grid(panel~., scale="free", labeller = label_parsed)

p = p + layer(data = df.top,  geom = c("point"), mapping = aes(color = c), alpha = 0.5, size = 4)
p = p + layer(data = df.bottom, geom = c("point"), mapping = aes(color = c), alpha = 0.5, size = 4)
p = p + theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust = 1))
p = p + guides(color = guide_legend(override.aes=list(size=6)))
p = p + scale_color_manual(values=getPalette(colourCount))

p = p + labs(color = "Phylum")

p.bact = p

p
```

```python
%%R -w 500 -h 400
DF = read.table("../data/BvP.l2fc.euk.csv", 
                sep = ",",
                header = TRUE)
df.plot = subset(DF, !is.na(DF$padj))

colourCount = length(unique(df.plot$Rank2))
getPalette = colorRampPalette(brewer.pal(5, "Dark2"))

df.top = data.frame(x = df.plot$Rank5, y = -log(df.plot$padj), OTU = df.plot$OTU, 
                    panel = "-log(p-value)", c = df.plot$Rank2)
df.bottom = data.frame(x = df.plot$Rank5, y = df.plot$log2FoldChange, OTU = df.plot$OTU,
                       panel = "log[2](planktonic:biofilm)", 
                       c = df.plot$Rank2)
d = rbind(df.top, df.bottom)
d.euk = d

p = ggplot(data = d, mapping = aes(x = x, y = y))
p = p + geom_hline(y = 0) 
p = p + facet_grid(panel~., scale="free", labeller = label_parsed)

p = p + layer(data = df.top,  geom = c("point"), mapping = aes(color = c), alpha = 0.5, size = 4)
p = p + layer(data = df.bottom, geom = c("point"), mapping = aes(color = c), alpha = 0.5, size = 4)
p = p + theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust = 1))
p = p + guides(color = guide_legend(override.aes=list(size=6)))
p = p + scale_color_manual(values=getPalette(colourCount))
p.euk = p

p
```

```python
%%R -w 1200
d.euk$libraryType = "Algae"
d.bact$libraryType = "Bacteria"
d = rbind(d.euk, d.bact)

colourCount = length(unique(d$c))
getPalette = colorRampPalette(brewer.pal(10, "Set1"))

FDR = -log(0.10)
OTUs.ltFDR.bact = subset(d.bact, (panel == "-log(p-value)")&(y >= FDR))$OTU
OTUs.ltFDR.euk = subset(d.euk, (panel == "-log(p-value)")&(y >= FDR))$OTU

p = ggplot(d, aes(x = x, y = y))
p = p + facet_grid(panel ~ libraryType, scales = "free", space = "free_x", labeller = label_parsed)
p = p + layer(data = d.euk,  geom = c("point"), mapping = aes(color = c), alpha = 0.65, size = 3)
p = p + layer(data = d.bact,  geom = c("point"), mapping = aes(color = c), alpha = 0.65, size = 3)
p = p + layer(data = subset(d.bact, (panel == "log[2](planktonic:biofilm)")&(OTU %in% OTUs.ltFDR.bact)),
              shape = 21, color = "black", size = 3, geom = c("point"))
p = p + layer(data = subset(d.euk, (panel == "log[2](planktonic:biofilm)")&(OTU %in% OTUs.ltFDR.euk)),
              shape = 21, color = "black", size = 3, geom = c("point"))
p = p + geom_hline(y = 0)
p = p + theme(axis.text.x = element_text(size = 7, angle = -90, vjust = 0.5, hjust = 0))
p = p + guides(color = guide_legend(override.aes=list(size=6)))
p = p + scale_color_manual(values=getPalette(colourCount))

p = p + labs(x = "", y = "", color = "Phylum")

p = p + guides(col = guide_legend(ncol = 2))

p = p + theme_bw()
p = p + theme(axis.text.x = element_text(angle = -90, vjust = 0.5, hjust = 0, size = 9.5), 
              axis.text.y = element_text(size = 14), 
              axis.title.x = element_text(size = 14), 
              axis.title.y = element_text(size = 14), 
              strip.text.x = element_text(size = 14), 
              strip.text.y = element_text(size = 14),
              legend.text = element_text(size = 12), 
              legend.title = element_text(size = 12), 
              strip.background = element_blank())

p.all = p

p
```

```python
%%R
ggsave("figs/l2fc_pval_all.pdf", p.all, w = 15, h = 8)
```
